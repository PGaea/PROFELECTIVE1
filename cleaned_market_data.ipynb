```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/PGaea/PROFELECTIVE1/refs/heads/main/market_data.csv')
```

```python
df.head()
```

```python
df['Item_Visibility'] = df['Item_Visibility'].replace(0, np.nan)
```

```python
df.head()
```

```python
df = df.drop(columns=['Outlet_Size'])
```

```python
df.head()
```

```python
df = df.fillna("N/A")
```

```python
df.head()
```

```python
df.to_csv("cleaned_market_data.csv", index=False)
```
